```python
import matplotlib.mlab as mlab
# the histogram of the data
mu=np.mean(X1)
sigma=np.std(X1)

print("mu = %f, sigma = %f" % (mu, sigma))

plt.figure(figsize=(10,10))
n, bins, patches = plt.hist(X1, 6, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ x\ for\ going\ straight}$')
plt.grid(True)

plt.show()
```
